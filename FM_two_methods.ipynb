```python
# Загружаем необходимые библиотеки

from scipy.io import wavfile
from scipy.fftpack import fft
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from IPython.display import Audio
import numpy as np

%matplotlib inline
```

```python
t = np.linspace(0,3,44100*3)
```

```python
sample_rate = 48000

sig = []

t00 = 0
t01 = 0

delta1 = 1/sample_rate
delta2 = 1/sample_rate

coef = 20

string1 = '101010010011110011101010101010101001010101010111111111'
  #        11101101011111001110101110101010110101010101011111111111111101100100100000010
string2 = '010101000101000010101010101111101101011111101010111010'
  #        0010101010111110101010

for i in range(0,len(string1),2):
    c = string1[i:i+2]
    if c == '00':
        delta = 1/sample_rate
    if c == '01':
        delta = 2/sample_rate
    if c == '10':
        delta = 3/sample_rate
    if c == '11':
        delta = 4/sample_rate
        
    c = string2[i:i+2]
    if c == '00':
        delta2 = 1/sample_rate
    if c == '01':
        delta2 = 2/sample_rate
    if c == '10':
        delta2 = 3/sample_rate
    if c == '11':
        delta2 = 4/sample_rate
        
    for j in range(sample_rate//coef):
        t00 = t00 + delta1
        t01 = t01 + delta2
        
        s = 5*np.sin(2*np.pi*440*t00)
        s += 5*np.sin(2*np.pi*550*t01)
        
        sig.append(s)
```

```python
Audio(data=sig,rate=sample_rate)
```
