```python
# Загружаем необходимые библиотеки

from scipy.io import wavfile
from scipy.fftpack import fft
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from IPython.display import Audio
import numpy as np

%matplotlib inline
```

```python
t = np.linspace(0,3,44100*3)
```

```python
sample_rate = 44100

sig = []

t0 = 0

delta = 1/sample_rate

coef = 100

string = '000101010010011110011101010101010101001010101010111111111'
         #000101010010011110011101010101101001010101010
for i in range(0,len(string),3):
    c = string[i:i+3]
    if c == '000':
        delta = 1/sample_rate
    if c == '001':
        delta = 2/sample_rate
    if c == '010':
        delta = 3/sample_rate
    if c == '011':
        delta = 4/sample_rate
    if c == '100':
        delta = 5/sample_rate
    if c == '101':
        delta = 6/sample_rate
    if c == '110':
        delta = 7/sample_rate
    if c == '111':
        delta = 8/sample_rate
        
    for j in range(sample_rate//coef):
        t0 = t0 + delta
        sig.append(5*np.sin(2*np.pi*200*t0))
```

```python
Audio(data=sig,rate=sample_rate)
```
